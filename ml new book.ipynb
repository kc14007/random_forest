```python
def univariable_Categorical(data,var):
    miss=data[var].isnull().sum
    uncat_cnt=data[var].nunique()
    uncat_list=data[var].unique()
    
    
    h=pd.DataFrame(car_sales[var].value_counts()).reset_index()
    h.columns=[var,"Count"]
    
    k=pd.DataFrame(car_sales[var].value_counts(normalize=True)).reset_index()
    k.columns=[var,"Percentage"]
    k["Percentage"]=(round(k["Percentage"]*100,2)).astype("str")+str("%")
    final=pd.merge(h,k, on=var , how="inner")
    
    print("Missing count: {miss}\n")
    print(f"Total unique count of categories: {uncat_cnt}\n")
    print(f"unique category list: n\{uncat_list}\n")
    
    plt.figure(figsize=(8,6))
    sns.countplot(data=data, x=var)
    plt.xticks(rotation=90)
    plt.show()
    return final
```

```python
univariable_Categorical(data=car_sales, var="Manufacturer")
```

```python
def univariate_numerical(data,var):
    results={"missing":data[var].min(),
             "max":data[var].max(),
             "sum":data[var].sum(),
             "average":data[var].mean(),
             "standard deviation":data[var].std(),
             "variance":data[var].var(),
             "skewness":data[var].skew(),
             "kurtosis":data[var].kurt(),
             "25th percentile":data[var].quantile(.25),
             "50th percentile":data[var].quantile(.50),
             "70th percentile":data[var].quantile(.75)}
    plt.subplot(121)
    sns.histplot(data[var])
    plt.subplot(122)
    sns.boxplot(data=data, y=var)
    plt.show()
    return results
             
```

```python
def missing(data):
    miss=pd.DataFrame(data.isnull().sum(), columns=["Missing_count"]).reset_index()
    miss["Missing_percentage"]=(round(miss["Missing_count"]/data.shape[0],3)*100).astype("str")+str("%")
    miss0=miss[miss["Missing_count"]>0].sort_values(by="Missing_count", ascending=False)
    return miss0
```
